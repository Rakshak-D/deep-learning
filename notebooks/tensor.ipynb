```python
def seperator(length:int=45):
    print("-"*length)
```

```python
print(f"Adding two tensors \n{x + x}")
seperator()
print(f"Multiplying two tensors \n{x * x}")
seperator()
print(f"Dividing two tensors \n{x // x}")
seperator()
print(f"Modulus two tensors \n{x % x}")
seperator()
print(f"Multiplying a tensor with a number \n{6 * x}")
```

```python
print(f"Transpose of x \n{tf.transpose(x)}")
seperator()
print(f"Concatenation (axis=0)\n {tf.concat([x,x],axis=0)}")
seperator()
print(f"Concatenation (axis=1)\n {tf.concat([x,x],axis=1)}")
```

```python
print(f"Soft max activation function\n{tf.nn.softmax(x,axis=-1)}")
seperator()
print(f"Sum of softmax activation function values\n{sum(tf.nn.softmax(x,axis=-1)[0])}") # Approx 1
```

```python
print(f"Sigmoid activation function\n{tf.nn.sigmoid(x)}")
```

```python
print(f"Sum of all values \n{tf.reduce_sum(x)}")
seperator()
print(f"Average of all values \n{tf.reduce_mean(x)}")
seperator()
print(f"AND operation on boolean values \n{tf.reduce_all([True,False])}") # Logical and
seperator()
print(f"OR operation on boolean values \n{tf.reduce_any([True,False])}") # Logical and
seperator()
print(f"Maximum of all values \n{tf.reduce_max(x)}") # Maximum value
seperator()
print(f"Product of all values \n{tf.reduce_prod(x)}") # Multiplies all value
```
